```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def what(S0, k, w, timestep, A=1, c=1):
    crossedzero = False
    t = 0
    time = [t]
    S_list = [S0]
    S = S0
    rho = -3*(1+w)
    print(f"The power of S:{2+rho}")
    for i in range(400):

        weird = A * (S ** (2+rho)) - k * (c**2)
        if weird < 1e-20:
            crossedzero = True
        if crossedzero:
            change = -np.sqrt(np.abs(weird)) * timestep
        else:
            change = np.sqrt(weird) * timestep       
        S += change 
        if S <0:
            break
        S_list.append(S)
        t += timestep
        time.append(t)
    return time, S_list
```

```python
#Q1:

k0 = what(.01, 0, 0, 0.01, A=1, c=1)
k1 = what(.01, 1, 0, 0.01, A=1, c=1)
k2 = what(.01, -1, 0, 0.01, A=1, c=1)

#plt.figure(figsize = (10,10))
plt.plot(k0[0], k0[1], label = 'k=0', color = 'g')
plt.plot(k1[0], k1[1], label = 'k=1', color = 'r')
plt.plot(k2[0], k2[1], label = 'k=-1', color = 'b')
plt.title("Matter dominated Universe")
plt.legend()
plt.ylabel('S(t)', fontsize = 15)
plt.xlabel('time')
plt.show()
```

```python
#Q2:

k0 = what(.01, 0, 1/3, 0.01, A=0.0002, c=0.1)
k1 = what(.01, 1, 1/3, 0.01, A=0.0002, c=0.1)
k2 = what(.01,-1, 1/3, 0.01, A=0.0002, c=0.1)

#plt.figure(figsize = (10,10))
plt.plot(k0[0], k0[1], label = 'k=0', color = 'g')
plt.plot(k1[0], k1[1], label = 'k=1', color = 'r')
plt.plot(k2[0], k2[1], label = 'k=-1', color = 'b')
plt.title("Radiation-dominated Universe")
plt.legend()
plt.ylabel('S(t)', fontsize = 15)
plt.xlabel('time')
plt.show()
```
